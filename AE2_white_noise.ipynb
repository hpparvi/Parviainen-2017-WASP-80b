# WASP-80b external data modelling: white noise

Here we fit the previously published WASP-80b light curves assuming *white noise*. The next step is to model the noise as a Gaussian process.

**Note:** The number of walkers, DE iterations, and MCMC iterations (as well as MCMC thinning and burn-in) have been chosen based on many trial runs. The idea of this notebook is to *just work*.

```python
%pylab inline 
from extcore import *
npop, niter_de, niter_mc = 150, 750, 10000
```

## 1a Constant radius ratio

Here we assume constant, wavelength independent, radius ratio.

```python
lpf = LPFTM(use_ldtk=False, constant_k=True, noise='white')
sampler_ck = Sampler('ckwn', lpf, npop, niter_de, niter_mc)
```

```python
sampler_ck.optimise()
sampler_ck.sample()
```

```python
sampler_ck.plot();
```

## 1b Constant radius ratio with LDTk

```python
lpf = LPFTM(use_ldtk=True, constant_k=True, noise='white')
sampler_ckld = Sampler('ckwn_ldtk', lpf, npop, niter_de, niter_mc)
```

```python
sampler_ckld.optimise()
sampler_ckld.sample()
```

```python
sampler_ckld.plot();
```

## 2a Wavelength-dependent radius ratio

```python
lpf = LPFTM(use_ldtk=False, constant_k=False, noise='white')
sampler_vk = Sampler('vkwn', lpf, npop, niter_de, niter_mc)
```

```python
sampler_vk.optimise()
sampler_vk.sample()
```

```python
sampler_vk.plot();
```

```python
lpf = LPFTM(use_ldtk=True, constant_k=False, noise='white')
sampler_vkld = Sampler('vkwn_ldtk', lpf, npop, niter_de, niter_mc)
```

```python
sampler_vkld.optimise()
sampler_vkld.sample()
```

```python
sampler_vkld.plot();
```
