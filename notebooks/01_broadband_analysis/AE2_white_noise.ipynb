# WASP-80b broadband analysis: white noise

This notebook carries out the joint analysis of the T13 and M14 broadband light curves assuming *white noise*. The next notebook does the same but includes also the F14 data and uses a Gaussian process to model the noise.

**Note:** The number of walkers, DE iterations, and MCMC iterations (as well as MCMC thinning and burn-in) have been chosen based on a set of trial runs. The idea of this notebook is to *just work*. The time to run all the simulations should be <1 h.

```python
%pylab inline
%run __init__.py
```

```python
from src.extcore import *
npop, niter_de, niter_mc = 150, 750, 10000
```

## 1a Constant radius ratio

Here we assume constant, wavelength independent, radius ratio.

```python
lpf = LPFTM(use_ldtk=False, constant_k=True, noise='white')
sampler_ck = Sampler(RFILE_EXT, 'ckwn', lpf, npop, niter_de, niter_mc)
```

```python
sampler_ck.optimise()
sampler_ck.sample()
```

```python
sampler_ck.plot();
```

## 1b Constant radius ratio with LDTk

```python
lpf = LPFTM(use_ldtk=True, constant_k=True, noise='white')
sampler_ckld = Sampler(RFILE_EXT, 'ckwn_ldtk', lpf, npop, niter_de, niter_mc)
```

```python
sampler_ckld.optimise()
sampler_ckld.sample()
```

```python
sampler_ckld.plot();
```

## 2a Wavelength-dependent radius ratio

```python
lpf = LPFTM(use_ldtk=False, constant_k=False, noise='white')
sampler_vk = Sampler(RFILE_EXT, 'vkwn', lpf, npop, niter_de, niter_mc)
```

```python
sampler_vk.optimise()
sampler_vk.sample()
```

```python
sampler_vk.plot();
```

## 2b Wavelength-dependent radius ratio with LDTk

```python
lpf = LPFTM(use_ldtk=True, constant_k=False, noise='white')
sampler_vkld = Sampler(RFILE_EXT, 'vkwn_ldtk', lpf, npop, niter_de, niter_mc)
```

```python
sampler_vkld.optimise()
sampler_vkld.sample()
```

```python
sampler_vkld.plot();
```
